```python
"""
This module calculates the normalized confusion matrices
of the matrix in input (in the confusion variable)
"""
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import regex as re

def get_f1_score(confusion_matrix, i):
    TP = 0
    FP = 0
    TN = 0
    FN = 0

    for j in range(len(confusion_matrix)):
        if (i == j):
            TP += confusion_matrix[i][j]
            tmp = np.delete(confusion_matrix, i, 0)
            tmp = np.delete(tmp, j, 1)

            TN += np.sum(tmp)
        else:
            if (confusion_matrix[i][j] != 0):

                FN += confusion_matrix[i][j]
            if (confusion_matrix[j][i] != 0):

                FP += confusion_matrix[j][i]

    recall = TP / (FN + TP)
    precision = TP / (TP + FP)
    # f1_score = 2 * 1/(1/recall + 1/precision)
    # F1 = 2TP / (2TP + FP + FN)
    f1_score = (2 * TP) / ((2 * TP) + FP + FN)
    recall = TP / (TP + FN)
    return f1_score, recall

def compute_scores(confusion):
    res = []
    rec = []
    for i in range(len(confusion)):
        f1, re = get_f1_score(confusion, i)
        res.append(f1)
        rec.append(re)
    print("macro f1:", np.mean(res)*100)
    print("balanced accuracy:", np.mean(rec)*100)

def comskg(confusion):    
    confusion_normalized = []
    for i in confusion:
        confusion_normalized.append(list(i / np.sum(i)))

    for i in confusion_normalized:
        st = "\\textit{angry} & "
        for j in i:
            st += str(round((j*100), 1)) + "// & "
        st = st[:len(st)-2]
        st += " \\\\ \hline"
        print(st)

    axis_labels = ["angry", "neutra", "sad", "happy", "fear", "disgust"]
    a_x = sns.heatmap(
                confusion_normalized, xticklabels=axis_labels, yticklabels=axis_labels,
                cmap='Blues', annot=True, fmt='.2f', square=True)
    plt.title("Confusion matrix")
    plt.ylabel("True label")
    plt.xlabel("Predicted label")
    plt.show()
```
